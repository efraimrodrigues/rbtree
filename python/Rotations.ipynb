```python
from node import node 
from rb_tree import rb_tree 
from utils import draw
import warnings

warnings.filterwarnings('ignore')

tree = rb_tree()

node_5 = node(5, 1, tree.get_nil(), tree.get_nil(), tree.get_nil())
node_6 = node(6, 0, node_5, tree.get_nil(), tree.get_nil())
node_8 = node(8, 0, node_6, tree.get_nil(), tree.get_nil())
node_9 = node(9, 0, node_6, tree.get_nil(), tree.get_nil())

node_10 = node(10, 0, node_8, tree.get_nil(), tree.get_nil())
node_11 = node(11, 0, node_8, tree.get_nil(), tree.get_nil())

node_5.set_right(node_6)
node_6.set_left(node_9)
node_6.set_right(node_8)

node_8.set_left(node_10)
node_8.set_right(node_11)

tree.root = node_5
```

```python
def left_rotate(tree, x):
    y = x.get_right()
    x.set_right(y.get_left())
    
    if y.get_left() != tree.get_nil():
        y.get_left().set_parent(x)
    y.set_parent(x.get_parent())
    
    if x.get_parent() == tree.get_nil():
        tree.set_root(y)
    elif x == x.get_parent().get_left():
        x.get_parent().set_left(y)
    else:
        x.get_parent().set_right(y)
    
    y.set_left(x)
    x.set_parent(y)
```

```python
draw(tree)
```

```python
left_rotate(tree, node_6)

draw(tree)
```

```python
def right_rotate(tree, x):
    y = x.get_left()
    x.set_left(y.get_right())
    
    if y.get_right() != tree.get_nil():
        y.get_right().set_parent(x)
    y.set_parent(x.get_parent())
    
    if x.get_parent() == tree.get_nil():
        tree.set_root(y)
    elif x == x.get_parent().get_right():
        x.get_parent().set_right(y)
    else:
        x.get_parent().set_left(y)
    
    y.set_right(x)
    x.set_parent(y)
```

```python
left_rotate(tree, node_6)
right_rotate(tree, node_8)
draw(tree)
```
